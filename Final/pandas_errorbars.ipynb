```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import sem
%matplotlib inline
```

```python
pd.options.display.mpl_style = 'default'
```

```python
np.random.seed(0)
s = pd.Series(np.random.randn(10))
```

### Original plot

```python
s.plot();
```

### Plot with error bars from list-like

```python
s.plot(yerr=np.ones(10));
```

### X and Y errors

```python
s.plot(yerr=np.ones(10), xerr=np.random.rand(10));
```

### Plot with error bars from DataFrame/dict

```python
d = {'x': np.random.randn(5), 'y': np.random.randn(5)}
df = pd.DataFrame(d)
d_err = {'x': np.random.rand(5), 'y': np.random.rand(5)}
df_err = pd.DataFrame(d_err)
```

```python
df.head()
```

```python
df_err.head()
```

```python
# This plotting method retrieves the error values for each data series
# a key-matched DataFrame/dict of error values
df.plot(yerr=df_err);
```

```python
# You can also specify the error values as a column in the plotting matrix
df.plot(y='y', yerr='x');
```

### Asymmetric error bars

```python
err = np.random.rand(2, 2, 5)*2
df.plot(yerr=err);
```

```python
df = pd.read_excel('/Users/robert/Downloads/OlympicAthletes_0.xlsx', 'OlympicAthletes')
```

```python
df.head()
```

```python
# group, average, and plot with error bars in only a few lines
sportgp = df.groupby('Sport')
agemean = sportgp.Age.mean().copy()
agestd = sportgp.Age.std()
agemean.sort(ascending=False)
agemean.plot(yerr=agestd, kind='bar', ecolor='0.3', figsize=(20, 4));
```

```python
df2 = df[[i in ['Basketball', 'Gymnastics', 'Equestrian'] for i in df.Sport]]
```

```python
sportgp2 = df2.groupby(('Sport', 'Year'))
agemean2 = sportgp2.Age.mean().unstack('Sport')
agesem2 = sportgp2.Age.agg(sem).unstack('Sport')
```

```python
agemean2.plot(yerr=agesem2);
```
